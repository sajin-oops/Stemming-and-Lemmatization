```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
text = "Hello there im sajin. nice to meet you"
words = word_tokenize(text)
print("------------Word Tokenize------------")
print("Before tokenize")
print(text)
print("\nAfter tokenize")

print(words)
```

```python
words = sent_tokenize(text)
print("------------sentence Tokenize------------")
print("Before tokenize")
print(text)
print("\nAfter tokenize")

print(words)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

# nltk.download('punkt')
# nltk.download('stopwords')
text = "i love python program. he is good at code"
tokens = word_tokenize(text)
stop_words = set(stopwords.words("english"))
filtered = [word for word in tokens if word.lower() not in stop_words]
print("Original Text")
print(text)
print("\nAfter word tokenize")
print(tokens)
print("\nAfter Stop words removal")
print(filtered)
```

```python
# Part Of Speech Tagging
import nltk
from nltk.tokenize import word_tokenize
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger_eng')
text = "Python is a good programming language"
tokens = word_tokenize(text)
result = nltk.pos_tag(tokens)
print("Original text")
print(text)
print("\nAfter performing Part Of Speech Tagging")
print(result)
```

```python
#Stemming
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
words = ["playing","played","joking","running"]
print("Real words")
print(words)
print("Performing Stemming\n")
for w in words:
    print(w,'->',stemming.stem(w))
    
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatization = WordNetLemmatizer()
words = ["playing","played","joking","running"]
for w in words:
    print(w , lemmatization.lemmatize(w))
    
```
